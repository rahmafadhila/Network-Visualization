# **Graph**

## **Nodes**

```python
# Kota-kota besar yang ada di jawa barat
kota_besar = ['Bandung',
              'Cianjur',
              'Sukabumi',
              'Bogor',
              'Purwakarta',
              'Karawang',
              'Bekasi',
              'Tasikmalaya',
              'Ciamis',
              'Banjar',
              'Sumedang',
              'Cirebon',
              'Indramayu',
              'Garut',
              'Kuningan',
              'Majalengka',
              'Subang']

display(len(kota_besar))
```

## **Nodes Position**

```python
kota_besar_position = [ (5 , 2),
                        (2.5 , 2), 
                        (1.5 , 1.5),
                        (1 , 3),
                        (4 , 3.3),
                        (3 , 3.9),
                        (2 , 4),
                        (8 , 0.8),
                        (9 , 0.8),
                        (10 , 0.5),
                        (7 , 2.4),
                        (10 , 2.8),
                        (9, 3.8),
                        (7, 1.2),
                        (9.5, 2),
                        (8.2, 2.5),
                        (6, 3.3) ]

len(kota_besar_position)
```

## **Edges for Nodes**

```python
# Ketergantungan antar kota
jalan_utama = [ ('Bandung', 'Sumedang'),
                ('Bandung', 'Tasikmalaya'),
                ('Bandung', 'Purwakarta'),
                ('Bandung', 'Cianjur'),
                ('Cianjur', 'Sukabumi'),
                ('Cianjur', 'Bogor'),
                ('Bogor', 'Sukabumi'),
                ('Bekasi', 'Karawang'),
                ('Karawang', 'Purwakarta'),
                ('Purwakarta', 'Cianjur'),
                ('Cirebon', 'Sumedang'),
                ('Tasikmalaya', 'Ciamis'),
                ('Ciamis', 'Banjar'),
                ('Indramayu', 'Cirebon'),
                ('Indramayu', 'Sumedang'),
                ('Purwakarta', 'Indramayu'),
                ('Karawang', 'Indramayu')]

display(len(jalan_utama))
```

## **Graph Nodes**

```python
G = nx.Graph()
for i in range(len(kota_besar)):
  G.add_node(kota_besar[i], pos=kota_besar_position[i])

# Plot kota-kota besar di Jawa Barat
position = nx.get_node_attributes(G, 'pos')
plt.figure(figsize=(12, 7))
nx.draw(G, position, with_labels=True, node_color='lightpink')
```

## **Graph Nodes with Edges**

```python
G.add_edges_from(jalan_utama)
plt.figure(figsize=(12, 7))
nx.draw(G, position, with_labels=True, node_color='lightpink', edge_color='darkred')
```

# **Degree Centrality**

```python
nx.degree_centrality(G)
```

```python
print(f'Kota terpenting berdasarkan degree centrality adalah {", ".join([i for i in nx.degree_centrality(G) if nx.degree_centrality(G)[i] == max(nx.degree_centrality(G).values())])}')
```

# **Betweenness Centrality**

```python
nx.betweenness_centrality(G)
```

```python
print(f'Kota terpenting berdasarkan betweenness centrality adalah {", ".join([i for i in nx.betweenness_centrality(G) if nx.betweenness_centrality(G)[i] == max(nx.betweenness_centrality(G).values())])}')
```

**Kesimpulan :** <br>
Jadi berdasarkan percobaan diatas, Bandung merupakan kota terpenting yang merepresentasikan jalur transportasi untuk menghubungkan kota-kota besar di Jawa Barat berdasarkan nilai degree centrality dan betweenness centrality.
